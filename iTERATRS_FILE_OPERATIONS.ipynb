```python
#generator special type of generator it uses keyword like yeild 
# example of generator for a cibe 

def cubes(n):
    for num in range(n):
        yield num**3

cubes(10)
```

```python
#map()
#suppose consider
#The map() function in Python is used to apply a given function to each element of an iterable (like a list, tuple, etc.) and return a new iterable (map object) with the results. 
#It helps perform the same operation on all elements without using a loop.


nums = [1, 2, 3, 4, 5, 6]

def square(x):
    return x**2

result = map(square, nums)

print(list(result)) 
# -----------------------------------------
```

```python
celsius = [0, 10, 20, 30, 40]

def to_fahrenheit(c):
    return (c * 9/5) + 32

result = map(to_fahrenheit, celsius)

print(list(result))
```

```python
# #A lambda function is a small, anonymous (nameless) function in Python used for simple operations. 
# It is defined using the lambda keyword and can take any number of arguments but only has one expression.
# It is often used where a function is needed for a short period, especially with functions like map(), filter(), and reduce().
```

```python
nums = [1, 2, 3, 4, 5, 6]

result = map(lambda x: x**2, nums)

print(list(result))
```

```python
#filter
#The filter() function in Python is used to select elements from an iterable (like list) based on a condition.
#It returns only those elements for which the given function returns True.
```

```python
# # regular funciton 
# def even_check(list1):
#     if num%2==0:
#         return True
# # num=8
# # even_check(num)

#using filter 
def even_check(num):
    return num % 2 == 0

nums = [1, 2, 3, 4, 5, 6, 7, 8]

result = filter(even_check, nums)

print(list(result))

```

```python
#The reduce() function in Python is used to apply a function cumulatively to all elements of an iterable,
#so that it reduces the entire sequence into a single final value
#. It is available in the functools module.

from functools import reduce

nums = [1, 2, 3, 4, 5]

result = reduce(lambda x, y: x * y, nums)

print(result)
```
