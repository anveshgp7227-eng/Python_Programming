```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

a = []

for row in matrix:
    a.append(row[1])

print(a)
```

```python
# in the above example it is like 3 lines of code 
# using list comprehension we can simply write it as follows 
a=[ row[1] for row in matrix]
a
```

```python
data = [[10, 20], [30, 40], [50, 60]]

out = [x for group in data for x in group]

print(out)
```
