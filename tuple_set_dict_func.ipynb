```python
# real world example 
# (order_id, customer_name, product, price, quantity)

orders = [
    (101, "Anvesh", "Laptop", 55000, 1),
    (102, "Rahul", "Phone", 20000, 2),
    (103, "Sara", "Headphones", 1500, 3),
    (104, "John", "Keyboard", 1200, 1)
]

# Total revenue calculation
total_revenue = 0

for order in orders:
    order_id = order[0]
    customer = order[1]
    product = order[2]
    price = order[3]
    quantity = order[4]

    total = price * quantity
    total_revenue += total

    print("Order ID:", order_id)
    print("Customer:", customer)
    print("Product:", product)
    print("Total Bill:", total)
    print("----------------------")

print("TOTAL REVENUE:", total_revenue)
```

```python
# know the difference between shallow copy and deep copy 
#shallow copy: 
import copy

org = {1, 2, 3, 4}
shallow = copy.copy(org)

shallow.add(12)

print("original:", org)
print("shallow:", shallow)
```

```python
#deep copy: original value not destryoed 
import copy

org2 = {1, 2, 3, 4}
deep = copy.deepcopy(org)

deep.add(12)

print("original:", org2)
print("deep:", deep)
```

```python
# deep copy: [1, 2, [99, 4]]
# 🔥 Key Idea
# deepcopy() creates a completely new object
# Even nested objects (like lists inside lists) are copied separately
# So changes in copy do NOT affect original at all
# ⚡ Difference (very important)
# 🔸 Shallow Copy
# Copies only outer structure
# Inner objects are shared
# 🔸 Deep Copy
# Copies everything recursively
# Fully independent object
```
