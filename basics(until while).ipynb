```python
# type casting(Chaingin from 1 data type to other )
# 2 types--> Implicit, Explicit
# IMplicit type casting --> 👉 Python automatically c
#converts one data type to another (usually to avoid data loss).
# Explicit--> 👉 You convert data types yourself using functions
#like int(), float(), str().
```

```python
#--> Implict 
a = 5      # int
b = 2.5    # float

c = a + b  # int automatically becomes float
print(c)   # 7.5
print(type(c))  # <class 'float'>
```

```python
#-->Explicit
a = "10"   # string
b = 5

c = int(a) + b   # manually converting string to int
print(c)   # 15
print(type(c))  # <class 'int'>
```
