```python
#The __init__ method runs automatically when an object is created.
#It is used to initialize object attributes.

class Student:
    def __init__(self, name, major):
        self.name = name
        self.major = major

s1 = Student("ash", "CS")
print(s1.name, s1.major)
```

```python
#⚙️ #Methods in a Class

#A method is a function inside a class.
#It uses self to access or modify object data.
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
            return
        self.balance -= amount

acct = BankAccount("ash", 100)
acct.deposit(50)
acct.withdraw(30)
print("Balance:", acct.balance)
```

```python
#🧾 __str__ vs __repr__
#__str__ → User-friendly output (used by print())
#__repr__ → Debug-friendly output (used in notebooks/interpreter)
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return f"{self.title} by {self.author}"

    def __repr__(self):
        return f"Book(title={self.title!r}, author={self.author!r})"

b = Book("Train Journey", "saloni")
print(str(b))
print(repr(b))
b
```
